```python
import re
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

```python
def read_queue_data(filename):
    timestamps = []
    queue_sizes = []

    with open(filename, 'r') as file:
        for line in file:
            parts = line.split(',')
            timestamp = int(parts[0].split(':')[1].strip()) / 1000
            queue_size = int(parts[1].split(':')[1].strip())
            timestamps.append(datetime.datetime.utcfromtimestamp(timestamp))
            queue_sizes.append(queue_size)

    return timestamps, queue_sizes

def plot_queue_sizes(timestamps, queue_sizes, title):
    plt.figure(figsize=(12, 6))
    plt.plot(timestamps, queue_sizes, linestyle='-', color='b')
    plt.xlabel('Timestamp')
    plt.ylabel('Queue Size')
    plt.title(title)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
filename = '/Users/imchengliang/Downloads/Distributed-System/Exercise-1/demo/target/classes/queue_log_1.txt'
timestamps, queue_sizes = read_queue_data(filename)
title = 'Server 1 Queue Size Over Time'
plot_queue_sizes(timestamps, queue_sizes, title)
```

```python
def read_turnaround_times(filename):
    turnaround_times = []
    pattern = r"turnaroundTime=(\d+)"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                turnaround_time = int(match.group(1))
                turnaround_times.append(turnaround_time)

    return turnaround_times

def plot_turnaround_times(turnaround_times):
    plt.figure(figsize=(20, 6))
    plt.plot(turnaround_times, linestyle='-', color='r')
    plt.xlabel('Request Number')
    plt.ylabel('Turnaround Time (ms)')
    plt.title('Turnaround Time per Request')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
filename = '/Users/imchengliang/Downloads/Distributed-System/Exercise-1/demo/target/classes/dataset/naive_server.txt'
turnaround_times = read_turnaround_times(filename)
plot_turnaround_times(turnaround_times)
```
